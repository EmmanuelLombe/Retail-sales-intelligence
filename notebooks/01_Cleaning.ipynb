```python
import pandas as pd 
df = pd.read_csv("C:/Users/emmlo/Downloads/Retail_Sales_Intelligence/data/retail_sales_dataset.csv")
df.head()
```

```python
#data frame information
df.info()
```

```python
#Number of null values within each column 
df.isnull().sum()
```

```python
#Dropping rows where every single column is null 
df.dropna(how='all', inplace=True)

print(df.shape)
print(df.isnull().sum())
```

```python
#filling in null values
df['age'] = df['age'].fillna(df['age'].median())
df['quantity']= df['quantity'].fillna(df['quantity'].median())
df['discount_pct']= df['discount_pct'].fillna(0)
df['days_to_ship']= df['days_to_ship'].fillna(df['days_to_ship'].median())
df['customer_satisfaction']= df['customer_satisfaction'].fillna(df['customer_satisfaction'].median())

print(df.isnull().sum())
```

```python
#Changing column data types 
df['age'] = df['age'].astype(int)
df['quantity']= df['quantity'].astype(int)
df['days_to_ship']= df['days_to_ship'].astype(int)
df['order_date'] = pd.to_datetime(df['order_date'], format='mixed', dayfirst=True)
df['return_flag'] = df['return_flag'].astype(bool)
df['customer_satisfaction']= df['customer_satisfaction'].astype(int)

print(df.dtypes)
```

```python
#Checking for any spelling or formatting inconsistencies 
print(df['order_date'].unique())
print("\n")
print(df['region'].unique())
print("\n")
print(df['city'].unique())
print("\n")
print(df['product_category'].unique())
print("\n")
print(df['product_name'].unique())
print("\n")
print(df['payment_method'].unique())
print("\n")
print(df['order_status'].unique())
print("\n")
```

```python
#Checking return flag for its unique values 
print(df['return_flag'].unique())
```

```python
#Checking the count for each customer satisfaction rating and the median 
print(df['customer_satisfaction'].value_counts())
print(df['customer_satisfaction'].median())
```

```python
#Capitalizing order status values to reduce spelling inconsistency
df['order_status']= df['order_status'].str.capitalize()
print(df['order_status'].unique())
```

```python
#dropping all duplicate rows
df = df.drop_duplicates()
print(df.duplicated().sum())
print(df.shape)
```

```python
#Checking the number of customers associated with different order quantities
df['quantity'].value_counts().sort_index(ascending=False).head(10)
```

```python
#checking the product category, sale amount, and unit price associated with quantity == 999
df[df['quantity'] == 999][['product_name', 'unit_price', 'sales_amount']]
```

```python
df = df[df['quantity'] != 999]
print(df.shape) 
```

```python
df.to_csv("C:/Users/emmlo/Downloads/Retail_Sales_Intelligence/data/retail_sales_cleaned.csv", index=False)
```

```python
#Transfering dataframe to postgre for connection to powerBI (utilised this route to be able to clean the file additionally and update postgre without having to manually add an updated csv file)
from dotenv import load_dotenv
import os
from sqlalchemy import create_engine

print(os.getcwd())

load_dotenv()

password = os.getenv("db_password")

engine = create_engine(f'postgresql://postgres:{password}@localhost:5432/Retail_sales')

df.to_sql(
    name="retail_sales_cleaned",
    con= engine,
    if_exists= "replace",
    index= False,
    method= "multi",
    chunksize= 1000
)
```
